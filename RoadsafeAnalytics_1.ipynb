```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
file_path ="/content/drive/MyDrive/US_Accidents_March23.csv"
```

```python
cols = ['City', 'State', 'Weather_Condition', 'Start_Time', 'End_Time', 'Severity']
df = pd.read_csv(file_path, usecols=cols, nrows=100000)

print(df.head())
print(df.info())
```

```python
print(f"\nDataset has {df.shape[0]} rows and {df.shape[1]} columns")

# List all columns with their data types
print("\nColumn Data Types:")
print(df.dtypes)

# More detailed info including non-null counts
print("\nDataset Info:")
df.info()
```

```python
missing_values =df.isnull().sum()
missing_percent = (missing_values / len(df) * 100)

missing_summary = pd.DataFrame({
    'Missing Values': missing_values,
    'Percent Missing': missing_percent
})

# Filter columns with missing data
missing_summary = missing_summary[missing_summary['Missing Values'] > 0].sort_values(by='Missing Values', ascending=False)

print("\nColumns with Missing Data:")
display(missing_summary)
```
